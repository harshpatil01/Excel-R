What is the probability that at least one in five attempted telephone calls reaches the wrong number? (Assume independence of attempts.)


#### Solution:

To solve this problem using Python, we can use the binomial distribution. 
The binomial distribution is a discrete probability distribution that describes 
the number of successes in a fixed number of independent trials, each with the 
same probability of success.

In this case, we want to know the probability that at least one in five attempted 
telephone calls reaches the wrong number, given that the probability of any one call 
being misdirected is 1/200. So, the number of trials is 5, the probability of success 
(i.e., a misdirected call) is 1/200, and we want to find the probability of at least one success.

We can use the cumulative distribution function (cdf) of the binomial distribution
to calculate this probability. The cdf gives us the probability of getting up to a certain number
of successes, so we subtract this from 1 to get the probability of getting at least one success. 

If 1 in 200 long-distance telephone calls are getting misdirected.  
Probability of call misdirecting   = 1/200
Probability of call not misdirecting = 1-1/200 = 199/200
The probability for at least one in five attempted telephone calls reaches the wrong number
Total number of trials(n) = 5
n = 5
Using Bernoulli Trial formula for calculating the probability,
p = 1/200
q = 1-p=199/200
P(x) = At least one in five attempted telephone calls reaches the wrong number
P(x) = (nCx)*(p^x)*(q^n-x)
P(x) = (nCx)* (p^x)* (q^n-x)    
P(1) = (5C1) *(1/200)^1* (199/200)^5-1
P(1) = 0.02475124687812502


The probability of atleast one call in five attempted phone calls reaching the wrong number
is 0.02475124687812502 or 24.7512468 %.

```python

#Code: 

from scipy.stats import binom

n = 5    # number of trials
p = 1/200   # probability of success
k = 0    # number of successes we want to exclude from the cdf

#Using binomial distribution function
prob = 1 - binom.cdf(k, n, p)

print(prob)
```

#### Q 3.	If X1 ~ N(μ, σ2) and X2 ~ N(μ, σ2) are iid normal random variables, then what is the difference between 2 X1 and X1 + X2? Discuss both their distributions and parameters.       


#### Solution:

To find the difference between 2X1 and X1 + X2, we can start by computing their distributions 
and parameters:

Let Y = 2X1 - (X1 + X2) = X1 - X2

Then, we have:

E(Y) = E(X1) - E(X2) = μ - μ = 0

Var(Y) = Var(2X1 - (X1 + X2)) = Var(X1) + Var(X2) - 2Cov(X1, X2) = σ^2 + σ^2 - 2σ^2 = 0

Therefore, Y has a degenerate distribution, which means that it takes on a single 
value with probability 1. In this case, that value is 0.

On the other hand, X1 + X2 has a normal distribution with mean μ + μ = 2μ and 
variance σ^2 + σ^2 = 2σ^2. Thus, X1 + X2 ~ N(2μ, 2σ^2).

In summary, the difference between 2X1 and X1 + X2 is a degenerate distribution with
a single value of 0, while X1 + X2 is a normal distribution with mean 2μ and variance 2σ^2.

We can solve this problem by using the standard normal distribution and then
transforming it back to the given normal distribution.

Let Z ~ N(0, 1) be the standard normal random variable.
Then, we can standardize the given normal random variable X as follows:

Z = (X - μ) / σ

where μ = 100 and σ = 20.

Substituting these values, we get:

Z = (X - 100) / 20

To find the values of a and b, we need to find the corresponding values of z1 and z2 such that:

P(z1 ≤ Z ≤ z2) = 0.99
The expression P(z1 ≤ Z ≤ z2) = 0.99 represents the probability that a 
standard normal random variable Z falls within the interval [z1, z2].

To solve for z1 and z2, we can use a standard normal distribution table or a calculator 
with a built-in normal distribution function.

For example, using a standard normal distribution table, we can find the z-scores 
corresponding to a cumulative probability of 0.99 in the table. This gives us:

P(Z ≤ z) = 0.99
z = 2.33

Note that this value of z represents the number of standard deviations above the 
mean that corresponds to a cumulative probability of 0.99.

Since the standard normal distribution is symmetric about the mean,
we can also find the z-score corresponding to a cumulative probability of 0.01
(which represents the area to the left of z1 or to the right of z2). This value is -2.33.

Therefore, we have:

z1 = -2.33
z2 = 2.33

So P(z1 ≤ Z ≤ z2) = 0.99 corresponds to the interval [-2.33, 2.33] for a standard normal distribution.


Using a standard normal distribution table, we can find that:

P(Z <= 2.33) = 0.9901

P(Z <= -2.33) = 0.009903075559164245

Therefore, we have:

P(-2.33 ≤ Z ≤ 2.33) = P[(a - 100) / 20 ≤ Z ≤ (b - 100) / 20] = 0.99

Solving for a and b, we get:


(a - 100) / 20 = -2.33 => a = 100 - 2.33 × 20= 53.4

(b - 100) / 20 = 2.33 => b = 100 + 2.33 × 20 = 146.6

Therefore, the two values symmetric about the mean, such that the probability of 
the random variable taking a value between them is 0.99, are a = 53.4  and b = 146.6 .

```python
"""
Questions 8 and 9 are based on the following: To prepare a report on the economy, analysts 
need to estimate the percentage of businesses that plan to hire additional employees in the 
next 60 days.

8.	How many randomly selected employers (minimum number) must we contact in order to guarantee
a margin of error of no more than 4% (at 95% confidence)?

A.	600
B.	400
C.	550
D.	1000


9.	Suppose we want the above margin of error to be based on a 98% confidence level. 
What sample size (minimum) must we now use?

A.	1000
B.	757
C.	848
D.	543



Solution:


To guarantee a margin of error of no more than 4% at a 95% confidence level, we can use the 
following formula:

n = (Z^2 * p * q) / E^2

where:
n = sample size
Z = z-score (corresponding to the confidence level, which is 1.96 for 95%)
p = proportion of businesses that plan to hire additional employees (unknown)
q = 1 - p
E = margin of error (0.04)

We don't know the value of p, so we can use 0.5 as a conservative estimate (since it maximizes
the sample size). Thus:

n = (1.96^2 * 0.5 * 0.5) / 0.04^2
n = 600.2499999999999

Therefore, we need to contact at least 600 employers to guarantee a margin of error of no more
than 4% at a 95% confidence level. The closest option is A, 600.

Answer: A. 600

    To achieve the same margin of error of no more than 4% but at a 98% confidence level,
    we need a larger sample size, since the z-score corresponding to 98% confidence level 
    is greater than that for 95%. The formula remains the same, except for the value of Z:

n = (Z^2 * p * q) / E^2

where:
n = sample size
Z = z-score (corresponding to the confidence level, which is 2.33 for 98%)
p = proportion of businesses that plan to hire additional employees (unknown)
q = 1 - p
E = margin of error (0.04)

Using the same conservative estimate of p = 0.5:

n = (2.33^2 * 0.5 * 0.5) / 0.04^2
n = 848.265625

Therefore, we need to contact at least 848 employers to guarantee a margin of error of no more 
than 4% at a 98% confidence level.

Answer: 848


"""



#Code

Z=1.96 #95 % confidence interval z score
p=0.5 #proportion of businesses that plan to hire additional employees (unknown)
q=1-p # proportion of businesses that plan not to hire additional employees (unknown)
E=0.04 #margin of error
n = (Z**2 * p * q) / E**2 #Sample Size
print('Therefore, we need to contact at least 600 employers to guarantee a margin of error of no more than 4% at a 95% confidence level.')

print('\n')

Z=2.33 #98 % confidence interval z score
p=0.5 #proportion of businesses that plan to hire additional employees (unknown)
q=1-p # proportion of businesses that plan not to hire additional employees (unknown)
E=0.04 #margin of error
n = (Z**2 * p * q) / E**2 #Sample Size
print('Therefore, we need to contact at least 848 employers to guarantee a margin of error of no more than 4% at a 98% confidence level.')

print('\n')
```
